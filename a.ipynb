```python
import pandas as pd
from sklearn.cluster import KMeans
from matplotlib import pyplot
from sklearn.preprocessing import MinMaxScaler
```

```python
data_path = 'Data/Uge15Data.csv'

temp_data = pd.read_csv(data_path)

data = pd.DataFrame(temp_data)


scaler = MinMaxScaler()
scaleddata = scaler.fit_transform(data)
scaleddataframe = pd.DataFrame(scaleddata)
kmeans = KMeans(n_clusters=3)

kmeans.fit(scaleddataframe)
y = kmeans.predict(scaleddataframe)



pyplot.scatter(scaleddataframe[0], scaleddataframe[1], c=y)
```
